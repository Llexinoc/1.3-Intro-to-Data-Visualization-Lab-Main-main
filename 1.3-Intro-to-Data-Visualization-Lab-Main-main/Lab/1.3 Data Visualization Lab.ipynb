The exec has decided that if you make a song that has all the most popular features of recent songs then it will definitely be a hit.  So if - for example - the most common key in the Spotify data is C# then we would want the song we are designing to also be in C#. **Hint:**  Import the Spotify data using `df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Spotify data
df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')

# Display the first few rows to understand the data
df.head()
```

```python
# Visualize BPM distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['bpm'], bins=30, kde=True)
plt.title('Distribution of BPM')
plt.xlabel('BPM')
plt.ylabel('Frequency')
plt.show()

# Describe BPM
print("BPM Statistics:")
print(df['bpm'].describe())
```

```python
# Visualize Key distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='key', order=df['key'].value_counts().index)
plt.title('Distribution of Keys')
plt.xlabel('Key')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Most common key
print("Most common key:", df['key'].mode()[0])
```

```python
# Visualize Mode distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='mode')
plt.title('Distribution of Modes')
plt.xlabel('Mode')
plt.ylabel('Count')
plt.show()

print("Most common mode:", df['mode'].mode()[0])
```

```python
# Visualize Danceability distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['danceability_%'], bins=20, kde=True)
plt.title('Distribution of Danceability')
plt.xlabel('Danceability (%)')
plt.ylabel('Frequency')
plt.show()

print("Danceability Statistics:")
print(df['danceability_%'].describe())
```

```python
# Visualize Energy distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['energy_%'], bins=20, kde=True)
plt.title('Distribution of Energy')
plt.xlabel('Energy (%)')
plt.ylabel('Frequency')
plt.show()

print("Energy Statistics:")
print(df['energy_%'].describe())
```

```python
# Visualize Speechiness distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['speechiness_%'], bins=20, kde=True)
plt.title('Distribution of Speechiness')
plt.xlabel('Speechiness (%)')
plt.ylabel('Frequency')
plt.show()

print("Speechiness Statistics:")
print(df['speechiness_%'].describe())
```

```python
# Visualize Acousticness distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['acousticness_%'], bins=20, kde=True)
plt.title('Distribution of Acousticness')
plt.xlabel('Acousticness (%)')
plt.ylabel('Frequency')
plt.show()

print("Acousticness Statistics:")
print(df['acousticness_%'].describe())
```

## Summary of Distributions and Criteria for a Smash Hit

Based on the visualizations and statistics:

- **BPM**: Normally distributed around 122, median 121. Typical range: 100-140.
- **Key**: Most common is C#.
- **Mode**: Major is more prevalent.
- **Danceability**: Mean 67%, median 69%. Typical range: 57-78%.
- **Energy**: Mean 64%, median 66%. Typical range: 53-77%.
- **Speechiness**: Mean 10%, median 6%. Typical range: 4-11%.
- **Acousticness**: Mean 27%, median 18%. Typical range: 6-43%.

For a guaranteed smash hit, we propose:
- BPM: 100-140
- Key: C#
- Mode: Major
- Danceability: 57-78%
- Energy: 53-77%
- Speechiness: 4-11%
- Acousticness: 6-43%

```python
# Find songs that match all criteria
criteria = (
    (df['bpm'] >= 100) & (df['bpm'] <= 140) &
    (df['key'] == 'C#') &
    (df['mode'] == 'Major') &
    (df['danceability_%'] >= 57) & (df['danceability_%'] <= 78) &
    (df['energy_%'] >= 53) & (df['energy_%'] <= 77) &
    (df['speechiness_%'] >= 4) & (df['speechiness_%'] <= 11) &
    (df['acousticness_%'] >= 6) & (df['acousticness_%'] <= 43)
)

matching_songs = df[criteria]

print("Songs that match all criteria:")
if not matching_songs.empty:
    for idx, row in matching_songs.iterrows():
        print(f"Artist: {row['artist(s)_name']}, Title: {row['track_name']}, Streams: {row['streams']}")
        # Check if popular: say if in_spotify_charts > 0 or streams > some threshold
        if row['in_spotify_charts'] > 0:
            print("  - This song was in Spotify charts.")
        else:
            print("  - This song was not in Spotify charts.")
else:
    print("No songs match all criteria.")
```

```python
# Pick a known smash hit: "Blinding Lights" by The Weeknd
smash_hit = df[df['track_name'].str.contains('Blinding Lights', case=False, na=False)]
if not smash_hit.empty:
    row = smash_hit.iloc[0]
    print(f"Features of 'Blinding Lights' by {row['artist(s)_name']}:")
    print(f"BPM: {row['bpm']}")
    print(f"Key: {row['key']}")
    print(f"Mode: {row['mode']}")
    print(f"Danceability: {row['danceability_%']}%")
    print(f"Energy: {row['energy_%']}%")
    print(f"Speechiness: {row['speechiness_%']}%")
    print(f"Acousticness: {row['acousticness_%']}%")
    print(f"Streams: {row['streams']}")
    
    # Compare to typical
    print("\nComparison to typical values:")
    print(f"BPM {row['bpm']} is within 100-140: {100 <= row['bpm'] <= 140}")
    print(f"Key {row['key']} is C#: {row['key'] == 'C#'}")
    print(f"Mode {row['mode']} is Major: {row['mode'] == 'Major'}")
    print(f"Danceability {row['danceability_%']}% is within 57-78: {57 <= row['danceability_%'] <= 78}")
    print(f"Energy {row['energy_%']}% is within 53-77: {53 <= row['energy_%'] <= 77}")
    print(f"Speechiness {row['speechiness_%']}% is within 4-11: {4 <= row['speechiness_%'] <= 11}")
    print(f"Acousticness {row['acousticness_%']}% is within 6-43: {6 <= row['acousticness_%'] <= 43}")
else:
    print("Blinding Lights not found in dataset.")
```
